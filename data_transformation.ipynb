```python
import lmdb
import os
import cv2
import json
```

```python
train_data_path = "../d/validation"

json_path = train_data_path + "/annos"
image_path = train_data_path + "/image"
```

```python
map_size_gb = 4
```

```python
lmdb_path = '../data/validation'
env = lmdb.open(lmdb_path, map_size=map_size_gb * (1024 **3))
```

```python
with env.begin(write=True) as txn:
    
    
    for json_file in os.listdir(json_path):        
        print(json_file)
        
        with open(os.path.join(json_path, json_file), 'r') as f:
            id  = json_file.split(".")[0]
            annotation = json.load(f)
            txn.put(f'annotation_{id}'.encode('utf-8'), json.dumps(annotation).encode('utf-8'))
    
    for image_file in os.listdir(image_path):
        print(image_file)
        id = image_file.split(".")[0]
        _, buffer = cv2.imencode('.jpg', cv2.imread(os.path.join(image_path, image_file)))
        txn.put(f'image_{id}'.encode('utf-8'), buffer.tobytes())
            
env.close()
```
